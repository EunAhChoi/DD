```python
p = 0.6
q = 0.5
```

```python
def getP_SA(p, E_H, E_T):
    print('p: {0}, E_H: {1}, E_T{2}'.format(p, E_H, E_T))
    P_SA = pow(p,E_H) * pow((1-p), E_T)
    print('P(S|A):',P_SA)
    return P_SA

def getP_SB(p, E_H, E_T):
    P_SB = pow(p,E_H) * pow((1-p), E_T)
    print('P(S|B):',P_SB)
    return P_SB
```

```python
def getP_AS(p, q, E_H, E_T):
    P_SA = getP_SA(p, E_H, E_T)
    P_SB = getP_SB(q, E_H, E_T)
    P_AS = P_SA / (P_SA + P_SB)
    P_AS = round(P_AS,2)
    print('P(A|S):',P_AS)
    return P_AS

def getP_BS(p, q, E_H, E_T):
    P_SA = getP_SA(p, E_H, E_T)
    P_SB = getP_SB(q, E_H, E_T)
    P_BS = P_SB / (P_SA + P_SB)
    P_BS = round(P_BS,2)
    print('P(B|S):',P_BS)
    return P_BS
```

```python
def get_A (p,q,E_H,E_T):
    P_AS = getP_AS(p, q, E_H, E_T)
    A_Head = (E_H*P_AS)
    A_Tail = (E_T*P_AS)
    print('number of A_Head:',A_Head)
    print('number of A_Tail:',A_Tail)
    return [A_Head, A_Tail]
    
    
def get_B (p,q,E_H,E_T):
    P_BS = getP_BS(p, q, E_H, E_T)
    B_Head = (E_H*P_BS)
    B_Tail = (E_T*P_BS)
    print('number of B_Head:',B_Head)
    print('number of B_Tail:',B_Tail)
    return [B_Head, B_Tail]
```

```python
E = {'E1':[5,5], 'E2':[9,1], 'E3':[8,2]}
```

```python
def updateP_Q(E, p, q):
    A_total=[0, 0]
    B_total=[0, 0]
    for E_num in E.keys():
        E_H = E[E_num][0]
        E_T = E[E_num][1]

        A_Head, A_Tail = get_A (p,q,E_H,E_T)
        A_total[0] = A_total[0] + A_Head
        A_total[1] = A_total[1] + A_Tail
        print()
        B_Head, B_Tail = get_B (p,q,E_H,E_T)
        B_total[0] = B_total[0] + B_Head
        B_total[1] = B_total[1] + B_Tail
        print('--------------------------------------\n')

    A_total[0] = round(A_total[0],2)
    A_total[1] = round(A_total[1],2)
    B_total[0] = round(B_total[0],2)
    B_total[1] = round(B_total[1],2)
        
    print('A total Head: {0}, Tail {1}'.format(A_total[0], A_total[1]))
    print('B total Head: {0}, Tail {1}'.format(B_total[0], B_total[1]))
    
    updated_p = round(A_total[0] / (A_total[0]+A_total[1]),2)
    updated_q = round(B_total[0] / (B_total[0]+B_total[1]),2)
    print('updated p : ', updated_p)
    print('updated q : ', updated_q)
    print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
    
    return [updated_p, updated_q]
    
```

```python
nowP = 0.1
nowQ = 0.9
count = 0
loop = True

while(loop):
    newP, newQ = updateP_Q (E, nowP, nowQ)
    if (nowP==newP) & (nowQ==newQ):
        print('End')
        print('Iteration count:', count)
        loop = False
    else:
        count = count +1
        nowP = newP
        nowQ = newQ
        
```
